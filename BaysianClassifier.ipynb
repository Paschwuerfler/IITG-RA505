```python

import numpy as np
from numpy import random
# Create a Gaussian Distribution function





persons = ["m", "m", "m", "m", "f", "f", "f", "f"]
height = [6, 5.92, 5.58, 5.92, 5, 5.5, 5.42, 5.75]
weight = [180, 190, 170, 165, 100, 150, 130, 150]
foot_size = [12, 11, 12, 10, 6, 8, 7, 9]

# Probability Tables 
# Because our data is continuous, we will use Gaussian Distribution
# Reminder: A normal probability table looks like this: 
#
# Outlook | yes | no | P(yes) | P(no) 
# --------+-----+----+--------+-------
#  Sunny  |  2  |  3 |  0.40  |  0.60
#  Rainy  |  3  |  2 |  0.60  |  0.40
#  Total  |  5  |  5 |  1.00  |  1.00


# Assuming 2 gaussian distributions (one for each Class (male/female))
# We have to calc mean and variance for each feature (height, weight, foot_size)

mean_height_male = np.mean([height[0:4]])
mean_height_female = np.mean([height[4:8]])
mean_weight_male = np.mean([weight[0:4]])
mean_weight_female = np.mean([weight[4:8]])
mean_foot_size_male = np.mean([foot_size[0:4]])
mean_foot_size_female = np.mean([foot_size[4:8]])

variance_height_male = np.var([height[0:4]])
variance_height_female = np.var([height[4:8]])
variance_weight_male = np.var([weight[0:4]])
variance_weight_female = np.var([weight[4:8]])
variance_foot_size_male = np.var([foot_size[0:4]])
variance_foot_size_female = np.var([foot_size[4:8]])

# Print the Data in a table 
print("\t Height mean  \tWeight mean  \tFoot Size mean")
print("-------------------------------------------------------")
print("M:\t" + str(mean_height_male) + "\t" + str(mean_weight_male) + "\t" + str(mean_foot_size_male))
print("F:\t" + str(mean_height_female) + "\t" + str(mean_weight_female) + "\t" + str(mean_foot_size_female))
print("")

# Print the variance in a table
print("\t Height variance  \tWeight variance  \tFoot Size variance")
print("-------------------------------------------------------")
print("M:\t" + str(variance_height_male) + "\t" + str(variance_weight_male) + "\t" + str(variance_foot_size_male))
print("F:\t" + str(variance_height_female) + "\t" + str(variance_weight_female) + "\t" + str(variance_foot_size_female))

```

```python
# doing an estimation 

test_height = 6
test_weight = 130
test_foot_size = 8

p_male = 0.5
p_female = 0.5

# calculating 

def wikinormal(x, mean, variance):
    return (1 / np.sqrt(2 * np.pi * variance)) * np.exp(-0.5 * ((x - mean)**2 / variance))

p_height_male = wikinormal(test_height, mean_height_male, variance_height_male)
print(p_height_male)

p_weight_male = wikinormal(test_weight, mean_weight_male, variance_weight_male)
print(p_weight_male)

p_foot_size_male = wikinormal(test_foot_size, mean_foot_size_male, variance_foot_size_male)
print(p_foot_size_male)

posterior_numerator_male = p_male * p_height_male * p_weight_male * p_foot_size_male
# Output the posterior probability
print("PM : " + str(posterior_numerator_male))

p_height_female = wikinormal(test_height, mean_height_female, variance_height_female)
print(p_height_female)

p_weight_female = wikinormal(test_weight, mean_weight_female, variance_weight_female)
print(p_weight_female)

p_foot_size_female = wikinormal(test_foot_size, mean_foot_size_female, variance_foot_size_female)
print(p_foot_size_female)

posterior_numerator_female = p_female * p_height_female * p_weight_female * p_foot_size_female

print("PF : " + str(posterior_numerator_female))


```
